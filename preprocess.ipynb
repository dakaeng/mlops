```python
import pandas as pd
import numpy as np

from library1 import *
```

```python
path = 'D:\KU\mlops\data'
train_chg = pd.read_csv(path + '/data/raw_data/train/1000_chg.csv')
train_dchg = pd.read_csv(path + '/data/raw_data/train/1000_dchg.csv')
test = pd.read_csv(path + '/data/raw_data/test/Test05_NG_chg.csv')
```

```python
# chg 데이터와 dchg 데이터 합치기
train = pd.concat([train_chg, train_dchg])
train = train.reset_index(drop = True)
```

```python
# 파이프라인 이용해서 전처리 (library1의 pipeline_model class에 속한 코드)
preprocess_pipe = Pipeline([
    ('feature_selector', VolTempSelector('M')),
    ('missing_value', handleMissingValue()),
    ('siff_smooth_df', DiffSmooth(0, 0, 0)),
    ('pca', PCA(n_components = 3)),
    ('component_selector', ComponentSelector(3))
])
```

```python
preprocess_pipe.fit_transform(train)
```

```python
preprocess_pipe.fit_transform(test)
```
